```python
import matplotlib.pyplot as plt
import math
```

```python
x = []

e=2.71828

for i in range(1,11):
    x.append(int(input("Enter Value of x : ")))    
    
```

```python
Mu = []
for j in range(1,11):
    Mu.append(float(input("Enter Value of Mueue(M) : ")))    
```

```python
l = len(x)

pos = []

print("   Value Of X   |   Value Of M (Mueue)  |  Poisson Probability Function")
print("====="*15)
for i in range(0,l):
#     for j in range(0,l):
    pos.append(  round(( (math.pow(Mu[i],x[i])) * (math.pow(e,-Mu[i])) ) / (math.factorial(x[i]) ),4)  )
    
    print("\t ",x[i],"\t|\t ",Mu[i]," \t \t|\t\t",pos[i])  
    
```

```python
import pandas as pd
import numpy as np

l = len(x)

pos = []


for i in range(0,l):
    for j in range(0,l):
        pos.append(  round(( (math.pow(Mu[j],x[i])) * (math.pow(e,-Mu[j])) ) / (math.factorial(x[i]) ),4)  )
    
pos = np.array(pos).reshape(10,10)

df = pd.DataFrame(pos,index=[x],columns=[Mu])
df
```

```python
import pandas as pd
import numpy as np

l = len(x)

pos = []


for i in range(0,l):
    for j in range(0,l):
        pos.append(  round(( (Mu[j]**x[i]) * (e**-Mu[j]) ) / (math.factorial(x[i]) ),4)  )
    
pos = np.array(pos).reshape(10,10)

df1 = pd.DataFrame(pos,index=[x],columns=[Mu])
df1
```

```python
plt.title("Bar Graph of Poisson Probability")
plt.xlabel("Value of X")
plt.ylabel("Poisson Probability")
plt.bar(x,pos)
```
