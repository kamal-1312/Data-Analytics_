```python
outliers = []
def detect_outliers_zscore(data):
  thres = 1.5
  mean = np.mean(data)
  std = np.std(data)
  print(mean,std)
  for i in data:
    z_score = (i-mean)/std
    if(np.abs(z_score)>thres):
      outliers.append(i)
  return outliers
sample = detect_outliers_zscore(mc['Pulse'])
print("Sample:",sample)
```

```python
outliers = []
def det(data):
  data = sorted(data)
  q1 = np.percentile(data,25)
  q3 = np.percentile(data,75)
  IQR = q3-q1
  l_bound = q1-(1.5*IQR)
  u_bound = q3+(1.5*IQR)
  for i in data:
    if i < l_bound or i > u_bound:
      outliers.append(i)
  return outliers

sample = det(mc['Pulse'])
print(sample)
```

```python
#print(sample)
for i in sample:
  a=np.delete(mc['Pulse'],np.where(mc['Pulse']==i))
  print(a)
```
